```python
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
import matplotlib.pyplot as plt


data = load_breast_cancer()
X = data.data
y = data.target
cv_scores =[]
depths = [1, 2, 3, 4, 5]

for d in depths:
  base_tree_classifier = DecisionTreeClassifier(max_depth= d, random_state=10)
  model = AdaBoostClassifier(base_tree_classifier, n_estimators=20, random_state=10)

  cross_val = cross_val_score(model, X, y, cv=10, scoring="accuracy")
  cv_scores.append(np.mean(cross_val))

plt.plot(depths, cv_scores, marker = "o")
plt.xlabel("Decision trees depths")
plt.ylabel("Cv_score per depths")
plt.title("Depths VS CV score")
plt.grid(True)
plt.show()



```
